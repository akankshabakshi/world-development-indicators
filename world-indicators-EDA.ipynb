```python
#let's explore indicators table
#1-selecting
indicators = pd.read_sql_query("""select * from indicators limit 5;""", conn)
```

```python
#view indicators table
#also side note, python index starts from 0 as we have specificied limit 5 - 0,1,2,3,4 will be shown
indicators
```

```python
#list the number of rows
pd.read_sql_query("""select count(*) as count from indicators;""",conn)
```

```python
#check for missing values in a column
pd.read_sql_query("""select count(*) as count_of_missing_values from indicators where indicatorname is null""", conn)
```

```python
#cecking for missing values in all column
pd.read_sql_query("""select count(*) - count(countryname) as missing_country_name,
                      count(*) - count(countrycode) as missing_country_code,
                      count(*) - count(indicatorname) as missing_indicator_name,
                      count(*) - count(indicatorcode) as missing_indicator_code,
                      count(*) - count(year) as missing_year,
                      count(*) - count(value) as missing_value from indicators;
""", conn)
```

```python
#4-JOINS
#let's make a join to get the information about indicators GDP related measures in 2014 in Euro area 
pd.read_sql("""select Indicators.*, Series.LongDefinition
                from Indicators
                left join series 
                on Indicators.IndicatorName = Series.IndicatorName
                where Indicators.IndicatorName like 'GDP%'
                and countryname = 'Euro area'
                and year = '2014'
                """, conn)
```

```python
#view records of indicators
pd.read_sql("""select * from indicators""", conn)
```

```python
#Find min value
pd.read_sql("""select min(value) from indicators""", conn)
```

```python
#Find max value
pd.read_sql("""select max(value) from indicators""", conn)
```

```python
#using when then case
pd.read_sql("""select *, 
               case when value <10000 then 'Low'
               when value>80000 then 'High'
               ELSE 'Medium' 
               END AS category
               from Indicators
               where IndicatorName = 'GDP per Capita (current US$)'
               and year = '2014'
                """, conn)
```

```python
#6-Window Function
#adding  the mean of last 3 years using window function 
pd.read_sql("""select *,
            ROUND(avg(value) over(partition by countrycode),0) as avg_3Y
            from Indicators
            where IndicatorName = 'GDP per capita (current US$)'
            and year in('2012', '2013', '2014')
                """, conn)
```

```python
#7-Common Table Expression
#A CTE will save results of a query temporary. It can help simplifying some queries and, for exammple, can also help filtering based on ranking results
#first place from each year

pd.read_sql("""with GDP_3 AS 
            (select *,
            rank() over(partition by year order by value desc) as myrank
            from Indicators
            where IndicatorName = 'GDP per capita(current US$)'
            and year in(2012, 2013,2014)
            )
            select * from GDP_3
            where myrank = 1;
            """, conn)
```
